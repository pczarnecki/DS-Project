```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
import scipy.constants as sp
import DataSetup as DS
import SADSProject as SA
```

Overall parameters: Using 60 spectral points, block size of 10k, eval-1 training set and eval-2 testing set errors recorded. 

```python
### Linear Regression
results = xr.open_dataset("DS_Linear.h5", engine = "netcdf4")

plt.semilogy(results.blocks, results.e_hist[0])
plt.semilogy(results.blocks, results.e_test[0])
plt.legend(["training", "testing"])
plt.xlabel("blocks (x 10,000 iterations)")
plt.ylabel("mean absolute RMSE")
plt.savefig("60Linear.png", bbox_inches='tight', dpi=600)
plt.show()
```

```python
### Ridge Regression
results = xr.open_dataset("DS_Ridge10-5.h5", engine = "netcdf4")

plt.semilogy(results.blocks, results.e_hist[0])
plt.semilogy(results.blocks, results.e_test[0])
plt.legend(["training", "testing"])
plt.xlabel("blocks (x 10,000 iterations)")
plt.ylabel("mean absolute RMSE")
plt.savefig("Ridge10-5.png", bbox_inches='tight', dpi=600)
plt.show()
```

```python
results = xr.open_dataset("DS_Lasso.h5", engine = "netcdf4")
```

```python
plt.semilogy(np.linspace(1, 6, 6), results.train_errors)
plt.semilogy(np.linspace(1, 6, 6), results.test_errors)
plt.legend(["training", "testing"])
x_ticks = ["$10^{-8}$, 152,196", "$10^{-7}$, 133,175", "$10^{-6}$, 50,291", "$10^{-5}$, 5,556", "$10^{-4}$, 75", "$10^{-3}$, 8"]
plt.xticks(np.linspace(1, 6, 6), labels = x_ticks, rotation = 'vertical')

plt.xlabel("(alpha, subset size)")
plt.ylabel("absolute RMSE")
plt.savefig("LASSO.png", bbox_inches='tight', dpi=600)
plt.show()
```
